### Data Cleaning 

```python
# checking for NaNs in the dataset
TFDA_2020.isnull().sum()
```

```python
# Convert 'TRDTYPE' to categorical (optional but useful)
TFDA_2020['TRDTYPE'] = TFDA_2020['TRDTYPE'].astype('category')
```

```python
TFDA_2020['TRDTYPE'].unique()
```

```python
# See the distribution of values in TRDTYPE
TFDA_2020['TRDTYPE'].value_counts(dropna=False)
```

```python
# Fill categorical columns with placeholders
TFDA_2020['USASTATE'] = TFDA_2020['USASTATE'].fillna('UNKNOWN')
TFDA_2020['DEPE'] = TFDA_2020['DEPE'].fillna('0000')
TFDA_2020['MEXSTATE'] = TFDA_2020['MEXSTATE'].fillna('OT')         # 'OT' = State Unknown
TFDA_2020['CANPROV'] = TFDA_2020['CANPROV'].fillna('OT')           # 'OT' = Province Unknown

```

```python
TFDA_2020['TRDTYPE'].value_counts(dropna=False)
```

```python
# 🛡 Step 1: Isolate all TRDTYPE == 2 rows (protect them)
trdtype_2_rows = TFDA_2020[TFDA_2020['TRDTYPE'] == 2]

# 🧹 Step 2: Clean the rest of the dataset (excluding TRDTYPE == 2)
other_rows = TFDA_2020[TFDA_2020['TRDTYPE'] != 2]
cleaned_other_rows = other_rows.dropna(subset=['DEPE', 'COMMODITY2', 'DF'])

# 🔁 Step 3: Combine cleaned rows and protected TRDTYPE==2 rows
TFDA_2020_cleaned = pd.concat([cleaned_other_rows, trdtype_2_rows], ignore_index=True)

# 🔄 Step 4: Reset index
TFDA_2020_cleaned.reset_index(drop=True, inplace=True)

# ✅ Step 5: Confirm TRDTYPE distribution
print(TFDA_2020_cleaned['TRDTYPE'].value_counts(dropna=False))

# 🔁 Optional: reassign to main variable
TFDA_2020 = TFDA_2020_cleaned
```

```python
# Fill numeric columns with zeros or neutral values
TFDA_2020['DF'] = TFDA_2020['DF'].fillna(0).astype(int)
TFDA_2020['COMMODITY2'] = TFDA_2020['COMMODITY2'].fillna(99).astype(int)  
```

```python
# Dropping duplicates 
# reset_index to reset the index after removing duplicates
TFDA_2020.drop_duplicates(inplace=True)
TFDA_2020.reset_index(drop=True)
```
